```python
#Importing our Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Reading our Data and Creating a DataFrame
Overdose_Deaths = pd.read_csv("VSRR_Provisional_Drug_Overdose_Death_Counts_20240930.csv")
Overdose_Deaths_df = pd.DataFrame(Overdose_Deaths)
```

```python
#Displaying our DataPoints
Overdose_Deaths_df
```

```python
#Removing any NaN values we find within our rows
Clean_Overdose_df = Overdose_Deaths_df.dropna()
print(Clean_Overdose_df)
```

```python
  # Drop rows where 'Indicator' contains 'Number of Drug Overdose Deaths' 
df_filtered = Clean_Overdose_df[Clean_Overdose_df['Indicator'] != 'Number of Drug Overdose Deaths']
```

```python
#Removing some columns we will not be needing
df_filtered.drop(columns=['Footnote', 'Footnote Symbol', 'Period'], inplace=True)
```

```python
df_filtered.info()
```

```python
#Reorganizing our Columns to be better alligned
df_filtered = df_filtered[['State', 'State Name', 'Year', 'Month', 'Indicator', 'Data Value', 'Percent Complete', 'Percent Pending Investigation', 'Predicted Value']]

df_filtered
```

```python
#Displaying our final version of our DataFrame
print(df_filtered)
```

```python
#Downloading our Data as a Json file for continued coding
df_filtered.to_json("Cleaned_Overdose_Data", orient="records", lines=True)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.transform import cumsum
from bokeh.palettes import Category20c
import pandas as pd
import numpy as np

# Grouping the data by 'Indicator' and summing up 'Data Value' for each category
overdose_data = df_filtered.groupby('Indicator')['Data Value'].sum()
overdose_data = overdose_data.dropna()
overdose_data = overdose_data.sort_values(ascending=False)

#  number of slices 
top_overdose_data = overdose_data.head(10)

# Convert data to a DataFrame and calculate angles
top_overdose_data_df = pd.DataFrame(top_overdose_data.reset_index(name='value'))
top_overdose_data_df['angle'] = top_overdose_data_df['value'] / top_overdose_data_df['value'].sum() * 2 * np.pi

# Calculate the percentage for hover display
top_overdose_data_df['percentage'] = (top_overdose_data_df['value'] / top_overdose_data_df['value'].sum()) * 100

# Use a palette that can handle more colors if needed (up to 20 categories with Category20c)
top_overdose_data_df['color'] = Category20c[len(top_overdose_data_df)]

# Create the pie chart figure
p = figure(height=600, width=600, title="Most Common Types of Overdoses in the US", toolbar_location=None,
           tools="hover", tooltips="@Indicator: @percentage{0.2f}%", x_range=(-0.5, 1.0))

# Add wedges (pie slices) based on the dynamic data
p.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend_field='Indicator', source=top_overdose_data_df)

# Output to an HTML file
output_file("dynamic_overdose_percentage_pie_chart.html")
p.legend.visible = False

# Hide axis and grid lines
p.axis.axis_label = None
p.axis.visible = False
p.grid.grid_line_color = None

# Show the plot
show(p)
```
